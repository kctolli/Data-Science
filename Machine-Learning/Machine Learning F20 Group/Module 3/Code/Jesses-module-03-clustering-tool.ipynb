```
import pandas as pd
from sklearn import cluster
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```
campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')
```

```
data = campaign.copy()
bins = [10,20,30,40,50,60,70,80,90, np.inf]
names = ['10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']
data['age_groups'] = pd.cut(data['age'], bins, labels=names)
```

```
#@title Columns to Consider
include_age = True #@param {type:"boolean"}
include_job = True #@param {type:"boolean"}
include_marital = True #@param {type:"boolean"}
include_education = True #@param {type:"boolean"}
include_default = True #@param {type:"boolean"}
include_housing = True #@param {type:"boolean"}
include_loan = True #@param {type:"boolean"}
include_contact = True #@param {type:"boolean"}
include_month = True #@param {type:"boolean"}
include_day_of_week = True #@param {type:"boolean"}
include_campaign = True #@param {type:"boolean"}
include_pdays = True #@param {type:"boolean"}
include_previous = True #@param {type:"boolean"}
include_poutcome = True #@param {type:"boolean"}
include_employment_variation_rate = True #@param {type:"boolean"}
include_consumer_price_index = True #@param {type:"boolean"}
include_consumer_confidence_index = True #@param {type:"boolean"}
include_euribor3m = True #@param {type:"boolean"}
include_employee_count = True #@param {type:"boolean"}
include_subscribed_to_term_deposit = True #@param {type:"boolean"}
filter_by_subscribed_to_term_deposit = False #@param {type:"boolean"}
selection = []
if include_age:
  selection.append("age_groups")
if include_job:
  selection.append("job")
if include_marital:
  selection.append("marital")
if include_education:
  selection.append("education")
if include_default:
  selection.append("default")
if include_housing:
  selection.append("housing")
if include_loan:
  selection.append("loan")
if include_contact:
  selection.append("contact")
if include_month:
  selection.append("month")
if include_day_of_week:
  selection.append("day_of_week")
if include_campaign:
  selection.append("campaign")
if include_pdays:
  selection.append("pdays")
if include_previous:
  selection.append("previous")
if include_poutcome:
  selection.append("poutcome")
if include_employment_variation_rate:
  selection.append("emp.var.rate")
if include_consumer_price_index:
  selection.append("cons.price.idx")
if include_consumer_confidence_index:
  selection.append("cons.conf.idx")
if include_euribor3m:
  selection.append("euribor3m")
if include_employee_count:
  selection.append("nr.employed")
if include_subscribed_to_term_deposit and\
 not filter_by_subscribed_to_term_deposit:
 selection.append("y")
X = pd.get_dummies(data[data.y=="yes"][selection], drop_first=True)\
 if filter_by_subscribed_to_term_deposit\
 else pd.get_dummies(data[selection], drop_first=True)
```

```
#@title Optimize clustering

minimum_number_of_clusters =  2#@param {type:"integer"}
maximum_number_of_clusters =  12#@param {type:"integer"}

# Altered from Machine Learning Pocket Reference 
inertias = []
sizes = range(minimum_number_of_clusters, maximum_number_of_clusters)
for k in sizes:
    k2 = KMeans(random_state=42, n_clusters=k)
    k2.fit(X)
    inertias.append(k2.inertia_)
fig, ax = plt.subplots(figsize=(6, 4))
pd.Series(inertias, index=sizes).plot(ax=ax)
ax.set_xlabel("K")
ax.set_ylabel("Inertia")
fig.show()
```

```
number_of_clusters =  5#@param {type:"integer"}
x_clusters = X.copy()
est = KMeans(n_clusters=number_of_clusters)
x_clusters["cluster"] = est.fit_predict(x_clusters)
all_clusters = [{column_name:value for column_name in X.columns for value in cluster} for cluster in est.cluster_centers_]
#cluster_dictionaries = {column_name: [cluster[column_name] for cluster in all_clusters] for column_name in X.columns}
pd.DataFrame(all_clusters)
```

```
#group = x_clusters.groupby(["cluster","y_yes"])
group = x_clusters.groupby(["cluster"])
group.std()
```
